```python
import numpy as np
from matplotlib import pyplot as plt
from cmath import *
```

```python
def sin_iteration(z,c = 1.0 + 1j, maxiter = 255):
  for i in range(maxiter):
    if abs(z.imag) > 50.0:
      return i
    else:
      z = c*sin(z)
  return i
```

```python
def cos_iteration(z, c=1j, maxiter = 255):
  for i in range(maxiter):
    if abs(z.imag) > 50.0:
      return i
    else:
      z = c*cos(z)
  return i
```

```python
def plot_cos_julia(npoints, c, ranges=[-2*np.pi, 2*np.pi, -2*np.pi, 2*np.pi], maxiter = 255):
   x = np.linspace(ranges[0], ranges[1] , npoints)
   y = np.linspace(ranges[2], ranges[3], npoints) * 1j
   X, Y = np.meshgrid(x, y)
   complex_space = X + Y
   data = np.zeros(complex_space.shape)
   for k, tab in enumerate(complex_space):
      for l, z in enumerate(tab):
         count = cos_iteration(z, c, maxiter)
         data[k][l] = count
   plt.figure(figsize=(16,12))
   plt.scatter(np.real(complex_space), np.imag(complex_space), c=data, cmap = 'inferno')
```

```python
def plot_sin_julia(npoints, c, ranges=[-2*np.pi, 2*np.pi, -2*np.pi, 2*np.pi], maxiter = 255):
   x = np.linspace(ranges[0], ranges[1] , npoints)
   y = np.linspace(ranges[2], ranges[3], npoints) * 1j
   X, Y = np.meshgrid(x, y)
   complex_space = X + Y
   data = np.zeros(complex_space.shape)
   for k, tab in enumerate(complex_space):
      for l, z in enumerate(tab):
         count = sin_iteration(z, c, maxiter)
         data[k][l] = count
   plt.figure(figsize=(16,12))
   plt.scatter(np.real(complex_space), np.imag(complex_space), c=data, cmap = 'inferno')
```
